```python
# import statements

from random import shuffle
from random import choice


global playing
playing=True

#class definitions

# class card has atributs(nubmer,sign) and method to return value of card according to
# the blacjack rule


class card:
    
    def __init__(self,number,sign):
        
        self.number=number
        self.sign=sign
        
    def value(self):
        
        if self.number in ['queen','jack','king']:
            value=10
        elif self.number =='ace':
            value=11
        else:
            value=self.number
        return value
    

# class desk has parent class card,when we initial desk we have to give instace deck 

class desk(card):
    
    def __init__(self,card):
        self.card=card
        
    def division(self):
        div=self.card[:2]
        for card in div:
            self.card.remove(card)   
        return div
    
    def hit(self):
        div=self.card[:1]
        for card in div:
            self.card.remove(card)   
        return div
    
    def suffle(self):
        
        shuffle(self.card)
        

#class diller has parent class desk,for initialing it is needed to give list of cards

class diller(desk):
    
    def __init__(self,name,card_list=None):
        
        if card_list is None:
            card_list = [] # create a new list
        self.card_list= card_list
        
        self.name=name
        
        self.card_list=card_list
        
    def add_division(self,division_list):
        for card in division_list:
            self.card_list.append(card)
            
    def result(self):
        numbers=[card.value() for card in self.card_list]
        
        while numbers.count(11)>=2:
            numbers.remove(11)
            numbers.append(1)   
        suma=0
        for n in numbers:
            suma=suma+n
        
        return suma

#class player has parent classes diller and desk

class player(diller,desk):
    
    def __init__(self,name,card_list=None,balance=100):
        
        self.balance=balance
        self.name=name
        if card_list is None:
            card_list = [] # create a new list
        self.card_list= card_list
       
        
    def change_balnce(self,value):
        
        if value>0:
            self.balance+=2*value
        else:
            self.balance+=value
            
#function definition

def deck_maker():
    
    ''' 
    generate deck of 52 cards,every card is instance of card class
    card has two atributes, value and sign
    one method that gives back value of card for calculation
    
    '''
    deck=[]
    j=0
    for s in ['heart','clube','diamond','spade']:
        for i in [2,3,4,5,6,7,8,9,10,'ace','queen','jack','king']:
            j+=1
            k="k"+str(j)
            k=card(i,s)
            deck.append(k)
    return deck

def player_bet(player):
    
    '''
    player inputs bet,if current balance is less then inputed bet
    function print current bet and again asks to input bet
    
    '''
    
    bet=int(input("{},please insert Your bet=".format(player.name)))
    while player.balance<bet:
        print('Your curren balanc is {})'.format(player.balance))
        bet=int(input("{},please insert Your bet=".format(player.name)))
        
    return bet


def display_all(player):
    
    '''
    displays all cards in card_list of palyer
    
    '''
    
    print()
    print("{} cards are:".format(player.name),end=" ")
    for card in player.card_list:
        print("{}:{}".format(card.sign,card.number),end=", ")
    print("\nScore is {}".format(player.result()))
    print()

def display_diller(diller):
    
    '''
    displays only second card of diller according to the rule
    
    '''
    
    card=diller.card_list[1]
    print("\n{} cards are:{}:{}".format(diller.name,card.sign,card.number))
    card=diller.card_list[1]
    print()
    
    
def hit_player(player,desk):
    
    '''
    asks player to hit or stop,and if palyer hits 
    call method of player to add card in player list
    '''
    hint=input("Please, insert h for hit,s for stand = ")
    if hint.lower()=='h':
        player.add_division(desk.hit())
    else:
        print("Score is {}".format(player.result()))
    return hint.lower()

def hit_diller(diller,desk):
    
    '''
    according to the rule of game when hit diller,
    method add_division is called until score is below 17
    
    '''
    
    while diller.result()<17:
        diller.add_division(desk.hit())
        
def player_bust(player):
    
    '''
    checking if player is over 21 score
    '''
    flag=False
    if player.result()>21:
        print("Sorry {}, You busts".format(player.name))
        flag=True
    return flag
    
def player_bust_check(player):
    
    '''
    checking if diller is over 21 score
    '''
    flag=False
    if player.result()>21:
        flag=True
    return flag

def black_jack(player,diller):
    '''
    checking if player or diller got black_jack
    '''
    
    flag=False
    
    if player.result()==21:
        print("Black Jack,{} has won ".format(player.name))
        flag=True
    elif diller.result()==21:
        print("Black Jack,{} has won ".format(diller.name))
        flag=True
    return flag
    
        
          
def winner(player,diller,bet):
    '''
    checking who is winner
    '''
    
    if player.result() > diller.result():
        player.change_balnce(bet)
        print('{} has won !'.format(player.name))
    elif player.result()< diller.result():
        player.change_balnce(-bet)
        print('{} has won !'.format(diller.name))
    else:
        print('Nobody won!!!')

def start_new_game(player,diller,bet):
    
    '''
    starts new game of player if it is possible
    '''
    if player.balance>2:
        start=input("If You wont to start new game, insert Y/N=")
        if start.lower()=='y':
            player.card_list=[]
            diller.card_list=[]
            start_game(player)
        else:
            print("Thank You,see You soon!!!")
        
    else:
        print('Dear player,no more mony,game is over!!!')
    
    
```

```python
def start_game(player):
    
    '''
    1.game starts with argument (player instace of class player)
    2.by default player has balance $100
    3.initialization  diller as inctance of  diller class
    4.initialization  desk as inctance of  desk class
    (for inizialization desk we need list of 52 card from card class)
    5.call method add_division that put two cards in card_list of player and diller
    6.next step is check if black_jack hapens
    if it happens variabila playing becomes false
    if black_jack does not happen then player decide if he want's to hit
    if player hit, player call method add_card to card_list of player and check result
    if result is over 21 playing =False
    if result of player is below 21 
    function hit diller is call, by the moment result is over 17
    if diller is over 21 playing is Fasle
    again call function hit_player
    and if answer is y add new card in card_list of player
    check if plyer burst
    if player choose n or plyer burst playing =Fasle
    after that we just check who wone
    if player won we add him bet to the balance
    if player lose we take bet from his balance
    7.player can chose if he wants to play
    if he has balance less the $2 game is over
    if he put bet greater the his balance he hes to put again 
    and than game can starts
    '''
    player1=player
    playing=True
    #formatin desk,diller
    bet=0
    diller1=diller("Diller")
    desk1=desk(deck_maker())
    desk1.suffle()
    
    #player insert bet
    while not( 2 < bet < 500 ):
        bet=player_bet(player1)
    
    #game starts with division of two cards
    player1.add_division(desk1.division())
    diller1.add_division(desk1.division())
    display_all(player1)
    display_diller(diller1)
    
    #checking blackjack
    
    if black_jack(player1,diller1)==False:
        if hit_player(player1,desk1)=="h":
            if player_bust(player1)==False:
                hit_diller(diller1,desk1)
                if player_bust(diller1):
                    playing=False     
            else:
                playing=False
        else:
            hit_diller(diller1,desk=desk1)
            player_bust(diller1)
            playing=False
    else:
        playing=False
        
            
    while playing:
    
        display_all(player1)
        display_all(diller1)
    
        if (hit_player(player1,desk1)).lower()=="s" or player_bust(player1) or player_bust(diller1):
            playing=False
    
    win=""
    
    if player_bust_check(player1):
        win="diller"
        print()
        print("{} has won!".format(diller1.name))
        display_all(diller1)
        player1.change_balnce(-bet)
        
    elif player_bust_check(diller1):
        win="player"
        print()
        print("{} has won!".format(player1.name))
        display_all(diller1)
        player1.change_balnce(bet)
        
    if win not in ["player","diller"]:
        winner(player1,diller1,bet=bet)
        
    start_new_game(player1,diller1,bet=bet)
    playing=False
        
```
